```python
import pandas as pd
from pathlib import Path
import hvplot.pandas
%matplotlib inline
```

```python
draft_df = pd.read_csv(Path('./draft-data-20-years.csv'),
                       index_col = 'DraftYear', 
                       parse_dates=False, 
                       infer_datetime_format=True
)
```

```python
draft_df.head()
```

```python
draft_df = draft_df[['Rk','Pk','Tm','Player','College','DraftYr','G','TOTPTS', 'TOTTRB', 'TOTAST', 'PPG']]
```

```python
last_5_yrs_draft_df = draft_df.loc[draft_df['DraftYr'] >= 2017]
```

```python
last_5_yrs_1_rnd_draft_df = last_5_yrs_draft_df.loc[last_5_yrs_draft_df['Pk'] <=30]
```

```python
group_a_df = last_5_yrs_1_rnd_draft_df.loc[last_5_yrs_1_rnd_draft_df['Pk'] <=5]
group_a_df.head()
```

```python
group_b_df = last_5_yrs_1_rnd_draft_df.loc[(last_5_yrs_1_rnd_draft_df['Pk'] >=13) & (last_5_yrs_1_rnd_draft_df['Pk'] <=17)]
```

```python
group_c_df = last_5_yrs_1_rnd_draft_df.loc[last_5_yrs_1_rnd_draft_df['Pk'] >=26]
```

```python
group_a_ave=group_a_df.groupby('DraftYear').mean().sort_values('DraftYear')
```

```python
group_a_ave
```

```python
group_a_ttl_g_plot = group_a_ave['G'].hvplot(
    title='Average 1st Round Top 5 NBA Picks Total Games Played 2017-2021',
    x='DraftYear',
    y='G',
    xlabel='Draft Year',
    ylabel='Total Games Played',
    label='Group-A'
)
```

```python
group_b_ave=group_b_df.groupby('DraftYear').mean().sort_values('DraftYear')
```

```python
group_b_ave
```

```python
group_b_ttl_g_plot = group_b_ave['G'].hvplot(
    title='Average 1st Round Middle 5 NBA Picks Total Games Played 2017-2021',
    x='DraftYear',
    y='G',
    xlabel='Draft Year',
    ylabel='Total Games Played',
    label='Group-B'
)
```

```python
group_c_ave=group_c_df.groupby('DraftYear').mean().sort_values('DraftYear')
```

```python
group_c_ave
```

```python
group_c_ttl_g_plot = group_c_ave['G'].hvplot(
    title='Average 1st Round Last 5 NBA Picks Total Games Played 2017-2021',
    x='DraftYear',
    y='G',
    xlabel='Draft Year',
    ylabel='Total Games Played',
    label='Group-C'
)
```
